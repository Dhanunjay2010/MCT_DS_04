```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (8,5)
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
print("First 5 Rows:")
print(df.head())
print("\nDataset Shape:", df.shape)
print("\nDataset Information:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df.drop("Cabin", axis=1, inplace=True)
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=[0, 12, 18, 30, 50, 80],
    labels=["Child", "Teen", "Young Adult", "Adult", "Senior"]
)
print("\nNew Features Created Successfully!")
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=25, kde=True)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
print("INSIGHT:")
print("Most passengers were between 20 and 40 years old.")
print("The number of elderly passengers was comparatively small.\n")
plt.figure(figsize=(6,4))
sns.barplot(data=df, x="Sex", y="Survived")
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()
print("INSIGHT:")
print("Female passengers had significantly higher survival rates than males.")
print("This suggests rescue efforts prioritized women.\n")
plt.figure(figsize=(6,4))
sns.barplot(data=df, x="Pclass", y="Survived")
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
print("INSIGHT:")
print("First-class passengers had the highest survival rate.")
print("Survival probability decreased as passenger class decreased.\n")
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Pclass", y="Fare")
plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
print("INSIGHT:")
print("First-class passengers paid much higher fares.")
print("Several outliers indicate some passengers paid exceptionally high ticket prices.\n")
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df,
    x="Age",
    y="Fare",
    hue="Survived",
    alpha=0.7
)
plt.title("Age vs Fare by Survival")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
print("INSIGHT:")
print("Passengers paying higher fares generally had better survival outcomes.")
print("Age alone does not strongly influence fare.\n")
plt.figure(figsize=(10,6))
numeric_df = df.select_dtypes(include=np.number)
corr_matrix = numeric_df.corr()
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Correlation Heatmap")
plt.show()
print("INSIGHT:")
print("Passenger class and fare are strongly related.")
print("Survival has a positive relationship with fare and a negative relationship with passenger class.\n")
sns.catplot(
    data=df,
    x="Pclass",
    y="Survived",
    col="Sex",
    kind="bar",
    height=5,
    aspect=1
)

plt.show()
print("INSIGHT:")
print("Females consistently had higher survival rates across all passenger classes.")
print("The class effect is especially noticeable among male passengers.\n")
print("="*50)
print("CONCLUSION")
print("="*50)
print("""
1. Most passengers were young adults.
2. Female passengers had much higher survival rates.
3. First-class passengers were more likely to survive.
4. Higher ticket fares were associated with better survival chances.
5. Passenger class played a major role in survival outcomes.

The dashboard successfully demonstrates data cleaning,
feature engineering, and visualization techniques using
the Titanic dataset.
""")
```
